```python
data = pd.read_csv("Revenue.csv", low_memory = False)
```

```python
# Remove the spaces in column names
data.columns = data.columns.str.replace(' ', '_')
```

```python
data.info()
```

```python
data['TOTAL_FUND'] = data['GENERAL_FUND'] + data['FEDERAL_FUND'] + data['CAPITAL_FUND'] + data['SPECIAL_FUND']
```

```python
data.head(4)
```

```python
rs = data[data['FISCAL_YEAR']==2019]
rs = rs.groupby(['CATEGORY'])['TOTAL_FUND'].sum().reset_index()
rs = rs.sort_values('TOTAL_FUND', ascending=False)
print(rs.to_string())
```
